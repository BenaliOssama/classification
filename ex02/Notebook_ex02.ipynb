```python
import numpy as np
import matplotlib.pyplot as plt

# Create 100 x values between -10 and 10
x = np.linspace(-20, 15, 100)

# Compute each sigmoid
sigmoid  = 1 / (1 + np.exp(-x))
sigmoid1 = 1 / (1 + np.exp(-(0.5*x + 3)))
sigmoid2 = 1 / (1 + np.exp(-(5*x  + 11 )))

# Plot all three on the same figure
plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid,  label="sigmoid")
plt.plot(x, sigmoid1, label="sigmoid1")
plt.plot(x, sigmoid2, label="sigmoid2")

# The 0.5 line — this is the decision boundary
# wherever the curve crosses this line, that's the threshold
plt.axhline(0.5, color='red', label="p = 0.5")

plt.legend()
plt.title("Sigmoid functions")
plt.xlabel("x")
plt.ylabel("probability")
plt.show()
```
