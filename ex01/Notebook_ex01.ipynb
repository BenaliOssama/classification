```python
from sklearn.linear_model import LogisticRegression

X = [[0],[0.1],[0.2], [1],[1.1],[1.2], [1.3]]
y = [0,0,0,1,1,1,0]

clf = LogisticRegression()
clf.fit(X, y)  # "learn from this data"

x_pred = [[0.5]]
print(clf.predict(x_pred))           # predicted class
print(clf.predict_proba(x_pred))     # probability for each class
print("Coefficient:", clf.coef_) # p(x) = 1 / (1 + e^(-(0.81786797 * x + (-0.87522391))))
print("Intercept:", clf.intercept_) # |^|
print("Score:", clf.score(X, y))
```
